```python
def check_df(data_path):
    # Veriyi yükleme
    dataframe = pd.read_csv(data_path)

    print("##################### Shape #####################")
    print(dataframe.shape)
    print("##################### Types #####################")
    print(dataframe.dtypes)
    print("##################### Head #####################")
    print(dataframe.head(3))
    print("##################### Tail #####################")
    print(dataframe.tail(3))
    print("##################### NA #####################")
    print(dataframe.isnull().sum())
    print("##################### Quantiles #####################")
    print(dataframe['Ratio'].quantile([0, 0.05, 0.50, 0.95, 0.99, 1]))

# Veri çerçevesini kontrol etme
data_path = "/female_ratio.csv"
check_df(data_path)
```

Shape (Şekil): Veri çerçevesinin boyutunu (satır, sütun) belirtir. Bu veri çerçevesinde 69 satır ve 2 sütun bulunmaktadır.

Types (Türler): Her sütunun veri türünü belirtir. "Entity" sütunu metin (object) türünde iken, "Ratio" sütunu ondalık sayı (float64) türündedir.

Head (Baş): Veri çerçevesinin ilk üç satırını gösterir. Bu başlangıç ​​satırları, veri çerçevesinin genel yapısını anlamak için kullanılır.

Tail (Kuyruk): Veri çerçevesinin son üç satırını gösterir. Bu son satırlar, veri çerçevesinin sonundaki verileri incelemek için kullanılır.

NA (Eksik Değerler): Her sütundaki eksik değerlerin sayısını belirtir. Bu veri çerçevesinde herhangi bir eksik değer bulunmamaktadır.

Quantiles (Çeyreklikler): "Ratio" sütununun çeyrekliklerini ve diğer yüzdelerini gösterir. Bu, veri dağılımının genel görünümünü anlamak için kullanılır. Örneğin, verinin %95'lik dilimindeki en yüksek oran 7.378'dir.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setini yükleme
data = pd.read_csv("veri_seti.csv")

# Gerekli önişlemleri yapma (ör. veri setini temizleme, dönüştürme, vb.)
# Örnek olarak, 'Ratio' sütununu sayısal veriye dönüştürelim
data['Ratio'] = pd.to_numeric(data['Ratio'], errors='coerce')

# Görseleştirme için sadece sayısal verileri içeren bir veri seti oluşturma
data_numeric = data.dropna()  # NaN değerlere sahip satırları kaldırma veya başka bir temizleme işlemi yapma

# Alt grafikler için figür ve eksenler oluşturma
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Veri seti üzerinde özelleştirilmiş görselleştirmeleri oluşturma
# Örneğin:
sns.histplot(data_numeric, ax=axs[0, 0], color='skyblue', bins=20)  # Histogram
axs[0, 0].set_title('Histogram')  # Başlık ekleme
axs[0, 0].set_xlabel('X-axis label')  # X eksenine etiket ekleme
axs[0, 0].set_ylabel('Y-axis label')  # Y eksenine etiket ekleme

sns.boxplot(data_numeric, ax=axs[0, 1], color='salmon')  # Kutu grafiği
axs[0, 1].set_title('Box Plot')  # Başlık ekleme
axs[0, 1].set_xlabel('X-axis label')  # X eksenine etiket ekleme
axs[0, 1].set_ylabel('Y-axis label')  # Y eksenine etiket ekleme

sns.kdeplot(data_numeric, ax=axs[1, 0], color='green', linestyle='--')  # Yoğunluk grafiği (KDE)
axs[1, 0].set_title('Density Plot')  # Başlık ekleme
axs[1, 0].set_xlabel('X-axis label')  # X eksenine etiket ekleme
axs[1, 0].set_ylabel('Y-axis label')  # Y eksenine etiket ekleme

sns.lineplot(data=data_numeric, ax=axs[1, 1], color='purple', marker='o')  # Çizgi grafiği
axs[1, 1].set_title('Line Plot')  # Başlık ekleme
axs[1, 1].set_xlabel('X-axis label')  # X eksenine etiket ekleme
axs[1, 1].set_ylabel('Y-axis label')  # Y eksenine etiket ekleme

# Grafikleri gösterme
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

def check_df(data_path):
    # Veriyi yükleme
    dataframe = pd.read_csv(data_path)

    print("##################### Shape #####################")
    print(dataframe.shape)
    print("##################### Types #####################")
    print(dataframe.dtypes)
    print("##################### Head #####################")
    print(dataframe.head(3))
    print("##################### Tail #####################")
    print(dataframe.tail(3))
    print("##################### NA #####################")
    print(dataframe.isnull().sum())
    print("##################### Quantiles #####################")
    print(dataframe[['0.1%', '0.25%', '0.5%', '1%', '10%', '5%']].quantile([0, 0.05, 0.50, 0.95, 0.99, 1]))

# Veri çerçevesini kontrol etme
data_path = "/content/2- share-of-women-in-top-income-groups.csv"
check_df(data_path)
```

Shape (Şekil): Veri setinizin boyutunu (satır ve sütun sayısı) gösterir. (168 satır, 8 sütun)

Types (Türler): Her bir sütunun veri tipini gösterir. "Entity" sütunu bir nesne (string) tipinde, "Year" sütunu bir tamsayı (int64) tipinde, diğer sütunlar ise ondalık sayı (float64) tipindedir.

Head (Baş): Veri setinizin ilk üç satırını gösterir. Bu, veri setinizin nasıl göründüğünü anlamanıza yardımcı olur.

Tail (Kuyruk): Veri setinizin son üç satırını gösterir. Bu, veri setinizin sonundaki verileri kontrol etmenize yardımcı olur.

NA (Eksik Değerler): Her sütundaki eksik değerlerin sayısını gösterir. Bazı sütunlarda (örneğin, "0.1%", "0.25%", "0.5%"), eksik değerler bulunmaktadır. Bu durum, ilgili yüzdelik dilimlerin bazı yıllar için veri eksikliği olduğunu gösterir.

Quantiles (Yüzdelikler): Belirli yüzdelik dilimlerinin ve belirli değerlerin yüzdelik dilimlerinin quantile değerlerini gösterir. Korelasyon:

Grafikte gösterilen veriler, korelasyon analizi için kullanılabilir. Korelasyon, iki değişken arasındaki ilişkinin gücünü ve yönünü gösteren bir istatistiksel ölçüdür. Bu durumda, iki değişken şunlardır:

Bağımsız değişken: Gelir dilimi
Bağımlı değişken: En zengin %1'lik dilimdeki kadınların oranı
Korelasyon analizi, gelir dilimi ile en zengin %1'lik dilimdeki kadınların oranı arasında pozitif bir korelasyon olduğunu gösterebilir. Bu, gelir dilimi arttıkça en zengin %1'lik dilimdeki kadınların oranının da arttığı anlamına gelir.

Korelasyon Katsayısı:

Korelasyon katsayısı, -1 ile 1 arasında bir değerdir. Korelasyon katsayısının:

0'a yakın olması: İki değişken arasında zayıf bir ilişki olduğunu gösterir.
1'e yakın olması: İki değişken arasında güçlü bir pozitif ilişki olduğunu gösterir.
-1'e yakın olması: İki değişken arasında güçlü bir negatif ilişki olduğunu gösterir.
Korelasyon analizi, aşağıdakiler gibi çeşitli amaçlar için kullanılabilir:

İki değişken arasındaki ilişkinin gücünü ve yönünü belirlemek
Bir değişkendeki değişimin diğer değişkeni nasıl etkileyeceğini tahmin etmek
Sebep ve sonuç ilişkilerini belirlemek

Korelasyon analizi yaparken dikkat edilmesi gerekenler:

Korelasyon, nedensellik anlamına gelmez. İki değişken arasında korelasyon olsa bile, bu birinin diğerine neden olduğu anlamına gelmez.
Korelasyon katsayısı, sadece iki değişken arasındaki ilişkinin gücünü ve yönünü gösterir. Diğer faktörleri de göz önünde bulundurmak önemlidir.
